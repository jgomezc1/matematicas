\begin{align*}
& \sigma_{ij,j}+f_i=0\\
& \sigma_{ij}=\sigma_{ji}
\end{align*}

\begin{align*}
& t_i^{(n)}=\sigma_{ij}n_j \\
& u_i=\overline{u}_i
\end{align*}

* $\sigma_{ij}$ es el tensor de esfuerzos.

* $f_i$ es el vector de fuerzas de cuerpo.

* $\overline{u}_i$ el vector de desplazamientos conocidos en $S_u$.

\begin{equation}
\sigma_{ij}=2\mu\varepsilon_{ij}+\lambda\varepsilon_{kk}\delta_{ij}
\end{equation}

y donde:

$$\varepsilon_{ij}=\frac12\left(u_{i,j}+u_{j,i}\right)$$
