##### Magnitud y normalización

Se sabe que el cuadrado de la norma esta dado por:

> $$\left\|a\right\|^2=\overrightarrow a\cdot\overrightarrow a$$

```python
asqr = a.dot(a)
mag_a = np.sqrt(asqr)
print(mag_a)
```

Mientras que el vector  normal se calcula de acuerdo con:

$$\widehat n\;=\frac{\overrightarrow a}{\left\|a\right\|}$$

```python
n = a/mag_a
vectors = [a , n]
plot_vector(vectors)
```
