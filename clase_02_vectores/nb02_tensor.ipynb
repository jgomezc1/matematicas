Un vector puede definirse como la asociaci√≥n de un escalar con cada direcci√≥n del espacio mediante una expresi√≥n que es homogenea y l√≠neal en los cosenos directores.

El escalar asociado con una direcci√≥n arbitraria $\mathbf{\hat{n}}$ con cosenos directores $n_1$ y $n_2$ esta dado por:


$$V_n = V_1 n_1 + V_2 n_2  $$

Para visualizar este escalar usamos

$$\mathbf{V_n} = \left\|\mathbf{V_n}\right\| \mathbf{\hat{n}} $$

el cual "vectoriza" la proyecci√≥n.

```python
vectors = []
V_1 = 2.0
V_2 = 1.0
m = np.arange(0.0,(1/2)*np.pi , np.pi/20)
#
for i in range(len(m)):
    n_1_vals = np.cos(m[i])
    n_2_vals = np.sin(m[i])
    proy = V_1*n_1_vals + V_2*n_2_vals
    proy = np.abs(proy)
    n = (proy*n_1_vals, proy*n_2_vals)
    vectors.append(n)   
plot_vector(vectors)
```

##### (! Atenci√≥n !En la visualizaci√≥n anterior las flechas no representan vectores. Estas solo se incluyen para indicar el valor absoluto de la proyecci√≥n en determinada direcci√≥n)


Grafiquemos ahora el sistema de referencia como infinitos vectores normales (unitarios).

y sobre cada direcci√≥n proyectemos el tensor usando:

$$\mathbf{T_n} = \mathbf{T_1} n_1 + \mathbf{T_2} n_2  $$

```python
vectors = []
m = np.arange(0.0 , 2*np.pi , np.pi/20)
sigma = [[4 , 1], [1 , 2]]
sigma = numpy.array(sigma)
for i in range(len(m)):
    n = (np.cos(m[i]) , np.sin(m[i]))
    t_n = sigma.dot(n)
    vectors.append(t_n)
plot_vector(vectors)
```
