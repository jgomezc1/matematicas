```python
def plot_refsystem():
    m = numpy.arange(0.0,2*numpy.pi , numpy.pi/20)
    vectors = []
    for i in range(len(m)):
        n = (numpy.cos(m[i]) , numpy.sin(m[i]))
        vectors.append(n)
    plot_vector(vectors)
plot_refsystem()
```
